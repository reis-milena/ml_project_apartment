# K-means

```python
wcss2 = []

for k in range(1,20):
    print(k)
    kmeans = KMeans(n_clusters=k,
                    random_state=0)
    kmeans.fit(x_var)
    wcss2.append(kmeans.inertia_)
```

```python
# elbow plot
elbow = px.line(x=range(1,20),y=wcss2) 
for i in range(0,len(wcss2)):
        elbow.add_hline(y=wcss2[i], line_width=1, line_dash="dash", line_color="black")
elbow.show()
```

k = 2

```python
from sklearn.metrics import silhouette_score

sil = []

# dissimilarity would not be defined for a single cluster, thus, minimum number of clusters should be 2
for k in range(2, 20):
  kmeans = KMeans(n_clusters = k, random_state=0).fit(x_var)
  labels = kmeans.labels_
  sil.append(silhouette_score(x_var, labels, metric = 'euclidean'))
```

```python
# silhouette plot
px.line(x=range(2,20),y=sil) 
```

```python
kmeans = KMeans(n_clusters=2,
                    random_state=0)
labels = kmeans.fit_predict(x_var)
```

```python
np.unique(labels,return_counts=True)
```

```python
px.scatter(x = x_var[:,3], #price
           y = x_var[:,5], #square_foot
           color=labels)
```
